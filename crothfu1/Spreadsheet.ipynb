```python
import csv
import os
import glob
import shutil
import pandas as pd

#all functions
def ocr_status(file_path, CardTypeID, status):
    # Create a temporary file to write the updated content
    temp_file_path = file_path + '_temp'

    # Open the input CSV file and the temporary file
    with open(file_path, 'r') as input_file, open(temp_file_path, 'w', newline='') as temp_file:
        reader = csv.reader(input_file)
        writer = csv.writer(temp_file)

        for row in reader:
            if len(row) >= 2 and row[1] == str(CardTypeID):
                row[2] = status
            writer.writerow(row)

    # Replace the original file with the updated content
    shutil.move(temp_file_path, file_path)

    print(f"CSV file '{file_path}' updated.")
    
    
def sort_csv_by_column(csv_file, sort_column):
    
    df = pd.read_csv(csv_file)  # Read the CSV file into a DataFrame
    df.sort_values([df.columns[sort_column]], axis=0, ascending=[False], inplace=True)  # Sort DataFrame by the chosen column
    
    # Write sorted data back to the CSV file
    df.to_csv(csv_file, index=False)

    
def compare_and_print(csv_file, array):
    # Read the CSV file
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)

    # Compare and print the array in the matching row
    for row in rows:
        if row[0] in array:
            row.extend([element for element in array if element != row[0]])

    # Write the updated rows back to the CSV file
    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)
        

def delete_rows(csv_file, start_row, end_row):
    # Read the CSV file
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        rows = list(reader)

    # Delete the specified range of rows
    if start_row < len(rows):
        del rows[start_row:end_row+1]

    # Write the updated rows back to the CSV file
    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)
        

def get_file_names(folder_path):
    file_names = []
    for file_path in glob.glob(os.path.join(folder_path, '*')):
        if os.path.isfile(file_path):
            file_names.append(os.path.basename(file_path))
    return file_names


def write_row(array, file_path):
    with open(file_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(array)
        
        
def write_column(array, file_path):
    with open(file_path, 'a', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(header)
        for item in array:
            writer.writerow([item])
        
        
def append_column(array, file_path, Offset):
    with open(file_path, 'a', newline='') as csvfile:
        writer = csv.writer(csvfile)
        for item in array:
            writer.writerow([None]*Offset+ [item])
```
